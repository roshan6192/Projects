```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/creditcard.csv')
credit_card_df.head()
```

```python
#separating the data for analysis

legit = credit_card_df[credit_card_df.Class == 0]
fraud = credit_card_df[credit_card_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transcations

credit_card_df.groupby('Class').mean()
```

```python
#UNDER_SAMPLING

#Build a sample dataset containing similiar distribution of normal transcations and fraudulent transcations

legit_sample = legit.sample(n=492)
```

```python
#Concatenate 2 dataframe

new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.Class.value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#spliting data into feature and targets

X = new_df.drop('Class', axis=1)
Y = new_df['Class']
```

```python
print(X.shape)
print(Y.shape)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, xtrain.shape, xtest.shape)
```

```python
print(Y.shape, ytrain.shape, ytest.shape)
```

```python
model = LogisticRegression()

model.fit(xtrain, ytrain)
```

```python
xtrian_prediction = model.predict(xtrain)
training_data_accuracy = accuracy_score(xtrian_prediction, ytrain)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
xtest_prediction = model.predict(xtest)
test_data_accuracy = accuracy_score(xtest_prediction, ytest)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
