```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading data set
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head(10)
```

```python
credit_card_data.info()
```

```python
credit_card_data.tail()
```

```python
#checking missing values 

credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction and fraudulant transaction

credit_card_data['Class'].value_counts()
```

```python
#separating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure for data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compares the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#Using Under Sampling
#Building a sample dataset containing similiar distribution of normal and fraudulant transcations.

legit_sample = legit.sample(n=492)
```

```python
#concate 2 dataFrames

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting data into features and targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split data into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Training the model : Logistic Regression

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
#Model evaluation

#Accuracy Score
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

```python
#Precision Score
from sklearn.metrics import precision_score
```

```python
#precision for training data
precision_train = precision_score(y_train, X_train_prediction)
print('Precision on training data : ', precision_train )
```

```python
#precision for test data
precision_test = precision_score(y_test, X_test_prediction)
print('Precision on test data : ', precision_test)
```

```python
#Recall Score
from sklearn.metrics import recall_score
```

```python
#recall for training data
recall_train = recall_score(y_train, X_train_prediction)
print('Recall on training data : ', recall_train )
```

```python
#recall for test data
recall_test = recall_score(y_test, X_test_prediction)
print('Recall on test data : ', recall_test)
```

```python
#F1 Score
from sklearn.metrics import f1_score
```

```python
#f1 score for training data
f1_train = f1_score(y_train, X_train_prediction)
print('F1 score on training data : ', f1_train )

#f1 score for test data
f1_test = recall_score(y_test, X_test_prediction)
print('F1 score on test data : ', f1_test)
```
