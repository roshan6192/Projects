```python
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/heart_disease_data.csv')
df.tail()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop('target', axis=1)
y = df['target']
print(x)
print(y)
```

```python
#x_train.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(12, activation='relu', input_dim=13, kernel_initializer='uniform'))
model.add(Dense(8, activation='relu', kernel_initializer='uniform'))
model.add(Dense(8, activation='relu', kernel_initializer='uniform'))
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=100, batch_size=20, verbose=1)
score = model.evaluate(x_test, y_test)
```

```python
model.metrics_names
```

```python
print('%s : %0.2f%%'% (model.metrics_names[1],score[1]*100))
```

```python
x1 = [[57,	1,	0,	130,	131,	0,	1,	115,	1,	1.2,	1,	1,	3	]]
x1 = np.array(x1)
print(x1)
```

```python
pred = model.predict(x1)
print(pred)
if pred[0][0]>0.5:
  print('Patient has heartdisease')
else:
  print("Patient is healthy")
```
